```python
import numpy as np
import open3d as o3d
```

```python
data_point = '00000'
npy_file_path = f'/home/digi2/colino_dir/detection_data/ermis_detection/detection_res_1/detected_bbox3d_{data_point}.npy'
data_np = np.load(npy_file_path, allow_pickle=True)
```

```python
data_np
```

```python
geometries = []
```

```python
detections_data = data_np.item()['detections']
detections_data
```

```python
bbox_pcds = []
bbox_boxes = []

for detections in detections_data:
    id = detections[0]
    label = detections[1]
    bbox_points = detections[2].astype(np.float64)
    bbox_centroid = detections[3].astype(np.float64)
    bbox_transform = detections[4].astype(np.float64)
    
    pcd = o3d.geometry.PointCloud()
    pcd.points = o3d.utility.Vector3dVector(bbox_points)
    bbox_pcds.append(pcd)
    
    bbox = o3d.geometry.OrientedBoundingBox.create_from_points(pcd.points)
    bbox.color = (1, 0, 0)
    bbox_boxes.append(bbox)
    
    
geometries += bbox_pcds
geometries += bbox_boxes
```

```python
origin_frame = o3d.geometry.TriangleMesh.create_coordinate_frame(size=1, origin=[0, 0, 0])
geometries += [origin_frame]
```

```python
o3d.visualization.draw_geometries(geometries)
```
